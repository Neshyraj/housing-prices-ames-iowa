```python
import pandas as pd
test = pd.read_csv('ATM_Test.csv')
test.head()
```

```python
#Create needed interactions
test['Weekday_Center'] = test['Weekday']*test['Center']
test['Shops_2'] = test['Shops']*test['Shops']
test['Shops_Center'] = test['Shops']*test['Center']
test['Downtown_Weekday'] = test['Downtown']*test['Weekday']
test['Shops_2_Center'] = test['Shops_2']*test['Center']
test['Shops_3'] = test['Shops']*test['Shops_2']
test['Shops_ATMs_Downtown'] = test['Shops']*test['ATMs']*test['Downtown']
test['Shops_Weekday_Center'] = test['Shops_Center']*test['Weekday']
```

```python
test.head()
```

```python
preds = ['Shops','ATMs','Weekday','Downtown','Center','High','Shops_2','Shops_Center','Downtown_Weekday',
         'Shops_2_Center','Shops_3','Weekday_Center','Shops_ATMs_Downtown','Shops_Weekday_Center']
```

```python
from sklearn.metrics import mean_squared_error
import pickle

loaded_model = pickle.load(open('my_model.h5', 'rb'))
predictions = loaded_model.predict(test[preds])
test_mse = mean_squared_error(predictions,test['Withdraw'])
print('Test MSE:',test_mse)
```
